```python
# importação bibliotecas

import pandas as pd                       # importação pandas
```

```python
colunas_filmes = ['star_rating', 'title', 'content_rating', 'genre', 'duration', 'actors_list']

file = './data/movies.csv'
filmes = pd.read_csv(file, header=0, names=colunas_filmes)

filmes.head(3)
```

```python
filmes.columns
```

```python
filmes[filmes.duration >= 100]
```

```python
# operadores logicos
# and (e)

# verdadeiro e verdadeiro
print(True and True)

# verdadeiro e falso
print(True and False)

# falso e verdadeiro
print(False and True)

# falso e falso
print(False and False)
```

```python
# or (ou)

# verdadeiro ou verdadeiro
print(True or True)

# verdadeiro ou falso
print(True or False)

# falso ou verdadeiro
print(False or True)

# falso or falso
print(False or False)
```

```python
# boas praticas
# use & no lugar de and
# use | no lugar de or
# use ! no lugar de not
# adcionar parenteses para evitar erros
# exemplo:
print(True and (True or False))
print((True and True) or False)
print(True and (True or False))
```

```python
# somente filmes do genero Drama e com duração maior que 200 minutos
dramas_longa_duracao = filmes[(filmes.duration >= 200) & (filmes.genre == 'Drama')]
dramas_longa_duracao
```

```python
# somente filmes do genero Drama e com duração maior que 200 minutos
dramas_longos = filmes[(filmes.duration >= 200) | (filmes.genre == 'Drama')]
dramas_longos
```

```python
# filmes do genero Crime, Drama, ou Ação
filmes_crime_drama_acao = filmes[(filmes.genre == 'Crime') | (filmes.genre == 'Drama') | (filmes.genre == 'Action')]
filmes_crime_drama_acao
```

```python
# utilizando o metodo isin
# filtro especificos
# filmes com avaliacao entre 9.0 e 9.3

filmes[filmes.star_rating.isin([9.0, 9.1, 9.2, 9.3])]
```

```python
# filtro com lista de strings
filmes[filmes.genre.isin(['Crime', 'Drama', 'Action'])]
```
