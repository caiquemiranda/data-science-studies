```python
# importação bibliotecas

import seaborn as sns                       # importação seaborn
import warnings                             # importação warnings

warnings.filterwarnings('ignore')           # ignora os warnings 
```

```python
# importando dataset tips do seaborn

tips = sns.load_dataset('tips')
tips.head(3)
```

```python
# importando dataset titanic do seaborn

titanic = sns.load_dataset('titanic')
titanic.head(3)
```

```python
# importando dataset flights do seaborn

fly = sns.load_dataset('flights')
fly.head(3)
```

```python
# pivot table

fly = fly.pivot_table(values='passengers', 
                       index='month', 
                       columns='year')

fly
```

```python
# stripplot

sns.stripplot(x='day', y='tip', data=tips);
```

```python
sns.stripplot(x='day', y='tip', data=tips, jitter=True);
```

```python
sns.swarmplot(x='day', y='tip', data=tips);
```

```python
# violin plot

sns.swarmplot(x='day', y='tip', color='black', data=tips);
sns.violinplot(x='day', y='tip', data=tips);
```

```python
# Correlação

corr = tips.corr()
corr
```

```python
# heatmap correlação

sns.heatmap(corr, annot=True);
```

```python
# mapa de calor correlações

sns.heatmap(fly);
```

```python
# clustermap

sns.clustermap(fly);
```

```python
# grafico lmplot
# simples regressão com os dados

sns.lmplot(x='total_bill', y='tip', hue='time', data=tips);
```
