```python
# importação bibliotecas

import seaborn as sns                       # importação seaborn
```

```python
# importando dataset titanic do seaborn

titanic = sns.load_dataset('titanic')
titanic.head(3)
```

```python
# importando dataset tips do seaborn

tips = sns.load_dataset('tips')
tips.head(3)
```

```python
# correlação
# joinplot(histogramas com scatter), relação entre variavel
# grafico linear


# grafico joinplot tips(totaltips :: tip)
sns.jointplot(x='total_bill', y='tip', data=tips);
```

```python
# grafico joinplot titanic(fare :: age)

sns.jointplot(x='fare', y='age', data=titanic);
```

```python
# grafico joinplot titanic(size :: tip)

sns.jointplot(x='size', y='tip', data=tips);
```
