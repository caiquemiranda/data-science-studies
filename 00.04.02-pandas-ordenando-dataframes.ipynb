```python
# importação bibliotecas

import pandas as pd                       # importação pandas
```

```python
colunas_filmes = ['star_rating', 'title', 'content_rating', 'genre', 'duration', 'actors_list']

file = './data/movies.csv'
filmes = pd.read_csv(file, header=0, names=colunas_filmes)

filmes.head(3)
```

```python
# ordenanação pelo titulo e retornando o dataframe
filmes.sort_values(by='title').head(6)
```

```python
# ordenando ascending = False, retorno descrescente
filmes.sort_values(by='title', ascending=False).head(6)
```

```python
# ordenação pr mais de um atributo
filmes.sort_values(by=['genre', 'duration']).head(10)
```

```python
# filtros
# criação lista com filmes maiores que 200 min

booldians = []

for length in filmes.duration:
    if length >= 200:
        booldians.append(True)
    else:
        booldians.append(False)
```

```python
booldians
```

```python
type(booldians)
```

```python
len(booldians)
```

```python
filmes.shape
```

```python
# criando serie para aplicar filtro no dataframe
duration = pd.Series(booldians)
duration.head(10)
```

```python
# aplicação do filtro
filmes[duration]
```

```python
# filtro de colunas
filmes[duration][['title', 'star_rating']]
```
