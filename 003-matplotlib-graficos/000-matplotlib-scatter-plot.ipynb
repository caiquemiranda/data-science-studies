```python
# grafico multi-linhas
x = []
y = []
y2 = []
y3 = []

for i in range(-50, 100, 5):
    x.append(i)
    y.append(i)
    y2.append(i * 2)
    y3.append(i + 50)

print(x)
print(y)
print(y2)
print(y3)
```

```python
plt.scatter(x, y, s=100, c='gray', marker='o')
plt.scatter(x, y2, s=30, c='purple', marker='^')
plt.scatter(x, y3, s=50, c='blue', marker='+');
```
