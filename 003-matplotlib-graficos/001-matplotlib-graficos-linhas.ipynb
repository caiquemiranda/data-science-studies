```python
x = []
y = []
y2 = []
```

```python
# grafico multi-linhas
x = []
y = []
y2 = []
y3 = []

for i in range(-50, 100, 5):
    x.append(i)
    y.append(i)
    y2.append(i * 2)
    y3.append(i + 50)

plt.plot(x, y, c='gray',ls=':' ,lw=1 ,marker='o')
plt.plot(y2, y3, c='purple',ls=':' ,lw=1 ,marker='^');
```
