```python
# importação bibliotecas

import matplotlib.pyplot as plt                       # importação matplotlib
```

```python
x = []
y = []
y2 = []
```

```python
for num in range(-200, 201, 10):
    x.append(num)
    y.append(num)
    y2.append(2*num)

print(x)

print(y)

print(y2)
```

```python
plt.plot(x, y, c='red', ls='-', lw=2, marker='o', ms=10, fillstyle='top', mfc='blue')
plt.plot(x, y2, c='green');
```

```python
# descrição de eixos

plt.plot(x, y, c='red', ls='-', lw=2, marker='o', ms=10, fillstyle='top', mfc='blue')
plt.plot(x, y2, c='green')

plt.xlabel('Eixo X');
```

```python
# descrição de eixos

plt.plot(x, y, c='red', ls='-', lw=2, marker='o', ms=10, fillstyle='top', mfc='blue')
plt.plot(x, y2, c='green')

plt.xlabel('Espaço para o eixo X', color='red', fontsize=15);
```

```python
# alinhamento

plt.plot(x, y, c='red', ls='-', lw=2, marker='o', ms=10, fillstyle='top', mfc='blue')
plt.plot(x, y2, c='green')

plt.xlabel('Espaço para o eixo X', color='red', fontsize=15, horizontalalignment='right');
```

```python
# eixo Y

plt.plot(x, y, c='red', ls='-', lw=2, marker='o', ms=10, fillstyle='top', mfc='blue')
plt.plot(x, y2, c='green')

plt.xlabel('Espaço para o eixo X', color='red', fontsize=15, horizontalalignment='center')
plt.ylabel('Espaço para o eixo Y', color='blue', fontsize=15, horizontalalignment='center');
```

```python
# eixo Y

plt.plot(x, y, c='red', ls='-', lw=2, marker='o', ms=10, fillstyle='top', mfc='blue')
plt.plot(x, y2, c='green')

plt.xlabel('Espaço para o eixo X', color='red', fontsize=15, horizontalalignment='center')
plt.ylabel('Espaço para o eixo Y', color='blue', fontsize=15, horizontalalignment='right');
```

```python
# titulo do grafico

plt.plot(x, y, c='red', ls='-', lw=2, marker='o', ms=10, fillstyle='top', mfc='blue')
plt.plot(x, y2, c='green')

plt.xlabel('Espaço para o eixo X', color='red', fontsize=15, horizontalalignment='center')
plt.ylabel('Espaço para o eixo Y', color='blue', fontsize=15, horizontalalignment='center')

plt.title('Titulo do grafico', color='gray', fontsize=25);
```

```python
# titulo do grafico

plt.plot(x, y, c='red', ls='-', lw=2, marker='o', ms=10, fillstyle='top', mfc='blue')
plt.plot(x, y2, c='green')

plt.xlabel('Espaço para o eixo X', color='red', fontsize=15, horizontalalignment='center')
plt.ylabel('Espaço para o eixo Y', color='blue', fontsize=15, horizontalalignment='center')

# numero fracionario
plt.title(r'Titulo do grafico $\frac{5}{4}$', color='gray', fontsize=20, horizontalalignment='center');
```

```python
# tickers

plt.plot(x, y, c='red', ls='-', lw=2, marker='o', ms=10, 
        fillstyle='top', mfc='blue', label='Primeiro')
plt.plot(x, y2, c='green', label='Segundo')

plt.xlabel('Espaço para o eixo X', color='red', fontsize=15, horizontalalignment='center')
plt.ylabel('Espaço para o eixo Y', color='blue', fontsize=15, horizontalalignment='center')

plt.title(r'Titulo do grafico $\frac{5}{4}$', color='gray', fontsize=20, horizontalalignment='center')

plt.legend(loc='lower right');

# customizando ticks X
eixoX = ['Janeiro', 'Abril', 'Junho']

# (valores atuais, novos valores)
plt.xticks([-200, 0, 200], eixoX);
```

```python
# tickers eixo Y

plt.plot(x, y, c='red', ls='-', lw=2, ms=10, 
        fillstyle='top', mfc='blue', label='Produto A')
plt.plot(x, y2, c='green',ls='-' ,label='Produto B')

plt.xlabel('Mês', color='red', fontsize=15, horizontalalignment='center')
plt.ylabel('Classificação', color='blue', fontsize=15, horizontalalignment='center')

plt.title('Produtos em 2018', color='gray', fontsize=20, horizontalalignment='center')

plt.legend(loc='lower right');

# customizando os ticks
eixoX = ['Janeiro', 'Abril', 'Junho']
eixoY = ['Fraco', 'Regular', 'Neutro', 'Bom', 'Excelente']

plt.xticks([-200, 0, 200], eixoX)
plt.yticks([-400, -100, 0, 100, 400], eixoY);
```
